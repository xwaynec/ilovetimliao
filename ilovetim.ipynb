```python
url = 'http://www.timliao.com/bbs/viewthread.php?tid=77092'
r = requests.get(
        url=url,
        headers={
            "Accept-Language": "zh-TW,zh;q=0.8,en-US;q=0.6,en;q=0.4",
        }
    )



soup = BeautifulSoup(r.text, 'html.parser', from_encoding="utf-8")
print soup.original_encoding
```
